```python
import numpy as np
import matplotlib.pyplot as plt
from pyunicorn.timeseries import RecurrencePlot

# Load time series from a text file
# Replace with the path to your own text file
file_path = r"E:\chaosgrandfinale\fnn-master\missed1.txt"
time_series = np.loadtxt(file_path)

# Create a RecurrencePlot instance
# Adjust the threshold parameter as needed
rp = RecurrencePlot(time_series, threshold=1)

# Compute RQA measures
recurrence_rate = rp.recurrence_rate()
determinism = rp.determinism(l_min=2)
average_diag_length = rp.average_diaglength(l_min=2)
laminarity = rp.laminarity(v_min=2)
trapping_time = rp.trapping_time(v_min=2)
entropy = rp.diag_entropy(l_min=2)

# Display the results
print(f"Recurrence Rate: {recurrence_rate}")
print(f"Determinism: {determinism}")
print(f"Average Diagonal Line Length: {average_diag_length}")
print(f"Laminarity: {laminarity}")
print(f"Trapping Time: {trapping_time}")
print(f"Entropy: {entropy}")

# Plot the recurrence matrix
plt.imshow(rp.recurrence_matrix(), cmap='binary', origin='lower')
plt.title('Recurrence Plot-net')
plt.xlabel('Time')
plt.ylabel('Time')
plt.show()
```
